## Embedding Techniques

converting text into vector

```python
import os
from dotenv import load_dotenv
load_dotenv()  # load all the env variables
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
```

```python
embedding = OpenAIEmbeddings(model="text-embedding-3-large")
# embedding = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024) # to set a perticular dim, default is 3072
embedding
```

```python
text = "this is tutorial on OPENAAI embedding"
query_result=embedding.embed_query(text)
query_result
```
